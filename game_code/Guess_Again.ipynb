## GUESS AGAIN 

Do you think you can beat the odds and guess better than random chance?   
If you can then try and see how well you can guess the average. 

```python
def guess_again(play1, play2, play3): #what are the possible plays you enter
    plays = list(range(1,10)) #possible plays 

    #storing all information for the human player 
    player1 = [play1, play2, play3] #empty list to store plays 
    player1_avg = (play1 +play2 +play3)/3 #average guess 

    #creating a random play from plays for player 2 
    import random as rd
    player2 = []
    for i in range(3): #loop that allow the selection of random values for computer and average        
        play_choice = rd.choice(plays)
        player2.append(play_choice)
    player2_avg = (player2[0] + player2[1] +player2[2])/3
    
    #creating random average 
    random_num = []
    for i in range(3): #loop that allow for comparison average        
        random_choice = rd.choice(plays)
        random_num.append(random_choice)
    random_choice_avg = (random_num[0] + random_num[1] + random_num[2])/3

    #criterion for winning the game:
    player1_score = []
    player2_score = []

    #Updating score for each player 
     #guessing exact average 
    if (player1_avg == random_choice_avg):
        player1_score.append(2)
    if (player2_avg == random_choice_avg):
        player2_score.append(2)

    #guessing within 1.5 of the average 
    if (player1_avg <= random_choice_avg +1.5  and player1_avg > random_choice_avg ) or (player1_avg >= random_choice_avg -1.5 and player1_avg < random_choice_avg ):
        player1_score.append(1)
    if (player2_avg <= random_choice_avg +1.5 and player2_avg > random_choice_avg ) or (player2_avg >= random_choice_avg -1.5 and player2_avg < random_choice_avg ):
        player2_score.append(1)

    #guessing within 2.5 of the average 
    if(player1_avg <= random_choice_avg +2.5  and player1_avg > random_choice_avg) or (player1_avg >= random_choice_avg -2.5 and player1_avg < random_choice_avg ):
        player1_score.append(0.5)
    if(player2_avg <= random_choice_avg + 2.5 and player2_avg > random_choice_avg) or (player2_avg >= random_choice_avg -2.5  and  player2_avg < random_choice_avg):
        player2_score.append(0.5)

    #guessing within 3.5 of the average 
    if(player1_avg > random_choice_avg +2.5 ) or (player1_avg < random_choice_avg -2.5 ):
        player1_score.append(0)
    if(player2_avg > random_choice_avg + 2.5 ) or (player2_avg < random_choice_avg -2.5 ):
        player2_score.append(0)
    
    print(f"Player 1 you have: {player1_score} points")

    
    
    


    
        


```
